# Gradient Descent Method

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

__Example)__ Solve the following problem:


\begin{matrix} minimize_{x, y} & x^2+y^2
\end{matrix}



```python
def f(x,y):
    return x**2+y**2
```

```python
f(1,2)
```

```python
xx=np.linspace(-50,50,100)# Array of 100 equal parts between -50 and 50
yy=np.linspace(-50,50,100)
x,y = np.meshgrid(xx,yy)# Coordinates Made by xx and yy
fig = plt.figure()
ax = fig.gca(projection='3d')
surf = ax.plot_surface(x,y,f(x,y),cmap='coolwarm')
```

The partial derivative of $f(x,y)$ is


$$\frac{\partial}{\partial x} f(x,y) = \frac{\partial}{\partial x} \big(x^2+y^2\big)= 2x$$

$$\frac{\partial}{\partial y} f(x,y) = \frac{\partial}{\partial y} \big(x^2+y^2\big)= 2y$$


Appling the gradient descent method, we can simplify update rules as follows:


$$x_{k+1}=x_k-\eta\frac{\partial}{\partial x} f(x_k,y_k)=x_k-\eta(2x_k)$$

$$y_{k+1}=y_k-\eta\frac{\partial}{\partial y} f(x_k,y_k)=y_k-\eta(2y_k)$$

Let $\eta=0.1$, # of iteration=20 and the initial point of $(x_0,y_0)=(1,1)$.

```python
eta=2 #Try another number 
iteration=100
x,y=1,1
```

```python
for i in range(iteration):
    
    df_dx=2*x
    df_dy=2*y
    
    x=x-eta*df_dx
    y=y-eta*df_dy
    print('Iteration:{}, (x,y)=({:.4f},{:.4f}), f(x,y)={:.4f}'.format(i,x,y,f(x,y)))
    
    
```

__Example)__ Solve the following problem:


\begin{matrix} minimize_{x, y} & \frac{-4x}{x^2+y^2+1}
\end{matrix}


```python
def g(x,y):
    return -4*x/(x**2+y**2+1)
```

```python
xx=np.linspace(-5,5,100)
yy=np.linspace(-5,5,100)
x,y = np.meshgrid(xx,yy)
fig = plt.figure()
ax = fig.gca(projection='3d')
surf = ax.plot_surface(x,y,g(x,y),cmap='coolwarm')
```

The partial derivative of $g(x,y)$ is

$$\frac{\partial}{\partial x} g(x,y) = \frac{4x^2-4y^2-4}{(x^2+y^2+1)^2}$$

$$\frac{\partial}{\partial y} g(x,y) = \frac{8xy}{(x^2+y^2+1)^2}$$

Appling the gradient descent method, we can simplify update rules as follows:

$$x_{k+1}=x_k-\eta\frac{4x_k^2-4y_k^2-4}{(x_k^2+y_k^2+1)^2}$$

$$y_{k+1}=y_k-\eta\frac{8x_ky_k}{(x_k^2+y_k^2+1)^2}$$

Let $\eta=0.1$, # of iteration=100 and the initial point of $(x_0,y_0)=(1,1)$.

```python
eta=0.01
iteration=1000
x,y=1,1
```

```python
for i in range(iteration):

    dg_dx=(4*x**2-4*y**2-4)/(x**2+y**2+1)**2
    dg_dy=(8*x*y)/(x**2+y**2+1)**2
    
    norm=np.sqrt(dg_dx**2+dg_dy**2)
    
    x=x-eta*dg_dx
    y=y-eta*dg_dy

    print('Iteration:{}, (x,y)=({:.4f},{:.4f}), g(x,y)={:.4f},Norm of g:{:.4f}'.format(i,x,y,g(x,y),norm))
```

# Linear Regression

Assume that $f(x_i)=ax_i+b$. Then the loss function $L(a,b)$ is


\begin{equation}
\begin{split}
L(a,b)
&=\frac{1}{N}\sum_{i=1}^N(y_i-f(x_i))^2\\
&=\frac{1}{N}\sum_{i=1}^N(y_i-ax_i-b)^2
\end{split}
\end{equation}

The partial derivative of $L(a,b)$ is

$$\frac{\partial}{\partial a}L(a,b)=-\frac{2}{N}\sum_{i=1}^Nx_i(y_i-ax_i-b)$$

$$\frac{\partial}{\partial b}L(a,b)=-\frac{2}{N}\sum_{i=1}^N(y_i-ax_i-b)$$

Appling the gradient descent method, we can simplify update rules as follows:

\begin{equation}
\begin{split}
a_{k+1}
&=a_k-\eta\frac{\partial}{\partial a}L(a_k,b_k)\\
&=a_k+\eta\frac{2}{N}\sum_{i=1}^Nx_i(y_i-a_kx_i-b_k)
\end{split}
\end{equation}

\begin{equation}
\begin{split}
b_{k+1}
&=b_k-\eta\frac{\partial}{\partial b}L(a_k,b_k)\\
&=b_k+\eta\frac{2}{N}\sum_{i=1}^N(y_i-a_kx_i-b_k)
\end{split}
\end{equation}

```python
raw_data = np.loadtxt('1_dim_data.csv',delimiter=",", skiprows=1) 
#separate data into x and y
x_data = raw_data[:,0]
y_data = raw_data[:,1]

plt.scatter(x_data,y_data)
```

__Step2) Initialize the variables__

```python
etha = 0.005
a, b = -1, -1
iteration = 1000

start_point = np.min(x_data)-10
end_point = np.max(x_data)+10
```

__Step3) Training__

```python
for i in range(iteration):
    Loss = np.sum((y_data - b - a*x_data)**2)/len(x_data)
    
    # learning process : using partial differential
    dL_db = (-2)*np.sum(y_data - b - a*x_data)/len(x_data)
    dL_da = 2*np.sum((y_data - b - a*x_data)*(-1)*(x_data))/len(x_data)
    b = b - etha*dL_db
    a = a - etha*dL_da
    
    norm=dL_db**2+dL_da**2
    
    print('Iteration:{}, Loss:{:.4f}, Norm of dL:{:.4f} a:{:.4f}, b:{:.4f}'.format(i,Loss,norm,a,b))


    
plt.scatter(x_data, y_data)
plt.plot([start_point,end_point],[a*start_point+b,a*end_point+b], c='r')
```

Since $L(a,b)$ is a convex function of $a$ and $b$, it has the minimum value at $\bigtriangledown L(a,b)=0$.

From $$\frac{\partial}{\partial{b}}L(a,b) = \frac{-2}{N} \sum_{i=1}^{N} (y_i - b - ax_i) = 0 \quad \rightarrow \quad b = \bar{y} - a \bar{x}$$

From $$\frac{\partial}{\partial{a}}L(a,b) = \frac{-2}{N} \sum_{i=1}^{N} x_i (y_i - b - a x_i) =0 \quad \rightarrow \quad a = \frac{\sum_{i=1}^{N} x_i y_i - N \bar{x}\bar{y}}{\sum_{i=1}^{N} {x_i}^2 - N {\bar{x}}^2}$$

where $\bar{x} = \frac{1}{N} \sum_{i=1}^{N} x_i$, $\bar{y} = \frac{1}{N}\sum_{i=1}^{N} y_i$

```python
N=len(x_data)
x_mean=np.sum(x_data)/N
y_mean=np.sum(y_data)/N

a=(np.sum(x_data*y_data)-N*x_mean*y_mean)/(np.sum(x_data**2)-N*x_mean**2)
b=y_mean-a*x_mean

print('a:{:.4f}, b{:.4f}'.format(a,b))
```

# Multiple Linear Regression

```python
raw_data = np.loadtxt('2_dim_data.csv',delimiter=",", skiprows=1) 
#separate data into x and y
x_data = raw_data[:,0]
y_data = raw_data[:,1]

plt.scatter(x_data,y_data)
```

Assume that $f(x_i)=a_2x_i^2+a_1x_i+a_0$. Then the loss function $L(a_2,a_1,a_0)$ is


\begin{equation}
\begin{split}
L(a_2,a_1,a_0)
&=\frac{1}{N}\sum_{i=1}^N(y_i-f(x_i))^2\\
&=\frac{1}{N}\sum_{i=1}^N(y_i-a_2x_i^2-a_1x_i-a_0)^2
\end{split}
\end{equation}

The partial derivative of $L(a_2,a_1,a_0)$ is

$$\frac{\partial}{\partial a_2}L(a_2,a_1,a_0)=-\frac{2}{N}\sum_{i=1}^Nx_i^2(y_i-a_2x_i^2-a_1x_i-a_0)$$

$$\frac{\partial}{\partial a_1}L(a_2,a_1,a_0)=-\frac{2}{N}\sum_{i=1}^Nx_i(y_i-a_2x_i^2-a_1x_i-a_0)$$

$$\frac{\partial}{\partial a_0}L(a_2,a_1,a_0)=-\frac{2}{N}\sum_{i=1}^N(y_i-a_2x_i^2-a_1x_i-a_0)$$

Appling the gradient descent method, we can simplify update rules as follows:

\begin{equation}
\begin{split}
a_2^{k+1}
&=a_2^k-\eta\frac{\partial}{\partial a_2}L(a_2^k,a_1^k,a_0^k)\\
&=a_2^k+\eta\frac{2}{N}\sum_{i=1}^Nx_i^2(y_i-a_2^kx_i^2-a_1^kx_i-a_0^k)
\end{split}
\end{equation}

\begin{equation}
\begin{split}
a_1^{k+1}
&=a_1^k-\eta\frac{\partial}{\partial a_1}L(a_2^k,a_1^k,a_0^k)\\
&=a_1^k+\eta\frac{2}{N}\sum_{i=1}^Nx_i(y_i-a_2^kx_i^2-a_1^kx_i-a_0^k)
\end{split}
\end{equation}

\begin{equation}
\begin{split}
a_0^{k+1}
&=a_0^k-\eta\frac{\partial}{\partial a_0}L(a_2^k,a_1^k,a_0^k)\\
&=a_0^k+\eta\frac{2}{N}\sum_{i=1}^N(y_i-a_2^kx_i^2-a_1^kx_i-a_0^k)
\end{split}
\end{equation}

Let $\eta=0.1$, # of iteration=1000 and the initial point of $(a_2,a_1,a_0)=(-1,-1,-1)$.

```python
etha = 0.1
a_2, a_1, a_0= -1, -1, -1
iteration = 1000

start_point = np.min(x_data)
end_point = np.max(x_data)
```

```python
for i in range(iteration):
    Loss = np.sum((y_data - a_2*(x_data**2)- a_1*x_data-a_0)**2)/len(x_data)
    
    dL_da_2 = (-2)*np.sum((x_data**2)*(y_data - a_2*(x_data**2) - a_1*x_data - a_0))/len(x_data)
    dL_da_1 = (-2)*np.sum(x_data*(y_data - a_2*(x_data**2) - a_1*x_data - a_0))/len(x_data)
    dL_da_0 = (-2)*np.sum(y_data - a_2*(x_data**2) - a_1*x_data - a_0)/len(x_data)
    
    a_2 = a_2 - etha*dL_da_2
    a_1 = a_1 - etha*dL_da_1
    a_0 = a_0 - etha*dL_da_0
    
    norm=dL_da_2**2+dL_da_1**2+dL_da_0**2
    
    print('Iteration:{}, Loss:{:.4f}, Norm of dL:{:.4f} a2:{:.4f}, a1:{:.4f}, a0:{:.4f}'.format(i,Loss,norm,a_2,a_1,a_0))


    
plt.scatter(x_data, y_data)
xx=np.linspace(start_point,end_point)
yy=[a_2*x**2+a_1*x+a_0 for x in xx]
plt.plot(xx,yy, c='r')
```
